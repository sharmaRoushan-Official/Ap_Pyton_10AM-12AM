# Copysign() function


 * __Returns a float consiting of the value of the first parameter and the second parameter.__

```python
import math
print(f'The copysign of these two numbers -3.14 and 2.718 is {math.copysign (-3.14,2.718)}')
print(f'The copysign of these two numbers 1729 and -0.577 is {math.copysign (1729,-0.577)}')
```

# factorial() function.

 * __It Returns the factorial of a numbers.__

```python
print(f'The factorial of number 3 is: {math.factorial(3)}') # 1x2x3=6
print(f'The factorial of number 8 is: {math.factorial(8)}')
```

# floor() function

 * __Rounds a number down to the nearest integer.__

```python
print(math.floor(3.9))
print(math.floor(10.20))
```

# gcd() function

 * __It Returns the greatest comman divisor of two strings.__

```python
print(math.gcd(3,10))
print(math.gcd(4,8))
```

# isqrt() functions

 * __Rounds a square root numbers downwards to the nearest integer.__
 * __The returned square root value is the floor value of square root of a non-negative integer numbers.__
 * __It gives a ValueError and TypeError when a negative integer number and a float number are used respectively.__

```python
print(math.isqrt(144))
print(math.isqrt(169))
print(math.isqrt(36))
```
