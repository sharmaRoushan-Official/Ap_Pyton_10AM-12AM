# Basic Data types of Python

```python
# String
print('Hello World')
print('Hello Python')
print('10')

# Integer

print(10)
print(1000)


# Float

print(45.8)
print(50.8)


# Boolean

print(True)
print(False)

print(bool(1))
print(bool(0))
```

# type() function

```python
# String
print(type('Hello world'))
print(type('100'))

# Integer
print(type(100))
print(type(-24))


#float

print(type(100.2))
print(type(-24.5))


# Boolean
print(type(True))
print(type(False))
```

# Converting an object type to another data type

```python
number=10

print(str(number))
print(type(str(number)))
print(type(number))

print(float(number))
print(type(float(number)))
```
